```python
#import neccessary ibraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
#read data from csv
df = pd.read_csv('subfiles/sub-data.csv')

#define neccessary variables 
x = 'AGE'
y = 'DEBT'

#Assign variables to correct axises in order to create scatterplot
plt.scatter(x = df[x], y = df[y])

#label Axises
plt.xlabel('age')
plt.ylabel('debt (in 100s of thousands)')

#Label entire scatter plot with a title 
plt.title('Age and Debt')

#Increase size of scatter plot 
plt.rcParams["figure.figsize"] = (30,20)
```
